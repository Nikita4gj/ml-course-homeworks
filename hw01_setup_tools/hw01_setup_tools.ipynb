```python
rng = np.random.default_rng(52)
X = rng.random((100, 5))*11-5

print(X)
```

```python
w = rng.random((5,))*11-5

print("The vector:", w, end="\n\n")

y = X @ w

print("The vectorized matrix:", y, sep="\n")
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)

df["target"]  = iris.target

names = iris.target_names

feature_names = iris.feature_names

df["species"] = names[df["target"]]

df.head()
```

```python
print(df.info())

df.describe()
```

```python
df.groupby("species")[feature_names].agg(["mean", "min", "max", "median"]).reset_index()
```

```python
sns.heatmap(df[feature_names].corr(), cmap="RdBu", vmin=-1, vmax=1)

plt.show()
```

```python
classes = df["species"].unique()

for cls in classes:

    subset = df[df["species"] == cls]

    plt.scatter(subset["petal length (cm)"], 
                subset["petal width (cm)"],
                label=cls, s = 20)

plt.legend()
plt.show()
```

```python
sns.scatterplot(
    data=df,
    x="petal length (cm)",
    y="petal width (cm)",
    hue="species"
)

plt.show()
```

```python
for name in names:
    fig, axes = plt.subplots(2,2, figsize=(10,8))
    for ax, feature in zip(axes.flatten(), feature_names):
        ax.hist(df[df["species"]==name][feature], bins=20, edgecolor="black")

        ax.set_title(feature)
```

```python
X = df[feature_names]

y = df["target"]

X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, stratify=y, random_state=52, train_size=0.8)
```

```python
model = sklearn.linear_model.LogisticRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Accuracy: {:.2%}".format(sklearn.metrics.accuracy_score(y_test, y_pred)))
```

```python
confusion_mat = sklearn.metrics.confusion_matrix(y_test, y_pred)

print(confusion_mat)

pd.DataFrame(confusion_mat, columns=names, index=names)
```
