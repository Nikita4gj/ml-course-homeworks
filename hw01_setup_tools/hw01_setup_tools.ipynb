```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mplt
import seaborn as sns
import sklearn
import sys 

print("Python version:", sys.version)
print("Pandas version:", pd.__version__)
print("Numpy version:",  np.__version__)
print("Sklearn version:",  sklearn.__version__)
```

```python
rng = np.random.default_rng(52)
X = rng.random((100, 5))*11-5

print(X)
```

```python
print(f"Mean of the entire matrix: {X.mean()}, by columns(asix = 0): {X.mean((0))}", sep="\t")
print(f"Std of the entire matrix: {X.std()}, by columns(asix = 0): {X.std(0)}", sep="\t")
print(f"Min of the entire matrix: {X.min()}, by columns(asix = 0): {X.min(0)}", sep="\t")
print(f"Max of the entire matrix: {X.max()}, by columns(asix = 0): {X.max(0)}", sep="\t")
```

```python
w = rng.random((5,))*11-5

print("The vector:", w, end="\n\n")

y = X @ w

print("The vectorized matrix:", y, sep="\n")
```

```python
data_fr = pd.DataFrame(X, columns=["sensor_1", "sensor_2", "sensor_3", "sensor_4", "sensor_5"])
print("Created data:")
data_fr
```

```python
data_fr["target"] = y
data_fr
```
