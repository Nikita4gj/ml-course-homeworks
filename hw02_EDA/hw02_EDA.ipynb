Разбил dataframe на числовые и нечисловые колонки, чтобы отдельно изучить каждые признаки

```python
missing = df.isnull().sum().sort_values(ascending=False)

missing[missing>0]
```

```python
for col in missing.index[:7]:
    print(col)
    print(df[col].value_counts(dropna=False))
    print()
```

Оценил пропуски в колонках PoolQC, MiscFeature и Alley буду удалять из-за слишком большого количество null и малоинформативности, остальные буду корректировать


```python
df.loc[~df["MasVnrType"].notnull(), "MasVnrArea"].value_counts(dropna=False)
```

Нашел аномалию для MasVnrArea, если type null то area может быть >0, следовалетьно тип скорее всего missing и надо это учесть в type а все остальное сделать none

```python
df.drop(missing.index[:3], axis=1, inplace=True)
```

```python
df.loc[~df["Fence"].notnull(), "Fence"] = "None"

df.loc[~df["FireplaceQu"].notnull(), "FireplaceQu"] = "None"

(~df.notnull()).sum().sort_values(ascending=False)
```

Удалил три первые по пропускам колонки PoolQC, MiscFeature, Alley и заменил null на None в Fence и FireplaceQu

```python
df.loc[~df["MasVnrType"].notnull() & df["MasVnrArea"]>0, "MasVnrType"] = "Missing"

df.loc[~df["MasVnrType"].notnull(), "MasVnrType"] = "None"

df.loc[~df["MasVnrArea"].notnull(), "MasVnrArea"] = 0

missing = (~df.notnull()).sum().sort_values(ascending=False)

missing[missing>0]
```
