Буду заполнять LotFrontage медианой, так как есть выбросы, а в GarageYrBlt занесу нули, так как года постройки гаража скорее всего нет, а так как есть HasGarage то все нормально

```python
df["LotFrontage"] =df["LotFrontage"].fillna(df["LotFrontage"].median())

df["GarageYrBlt"] = df["GarageYrBlt"].fillna(0)

plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
sns.boxplot(x= df["LotFrontage"])
plt.title("LotFrintage boxplot")

plt.subplot(2, 1, 2)
sns.boxplot(x = df["GarageYrBlt"])
plt.title("GarageYrBlt")

plt.tight_layout()

```

```python
missing = df.isnull().sum()

missing = missing[missing>0]

num_cols = df.select_dtypes(include="number")

num_cols.info()
```

```python
for col in missing.index:
    if col == "Electrical":
        df[col] = df[col].fillna(df[col].mode()[0])
    else:
        df[col] = df[col].fillna("None")

missing = df.isnull().sum()

missing = missing[missing>0]

missing.empty
```

```python
num_cols = df.select_dtypes(include="number")

num_info = num_cols.describe()

```

```python
new_info = pd.DataFrame(index=["mode", "var", "skew", "kurt", "q05", "q95"], columns=num_cols.columns)

for col in new_info.columns:
    mode = num_cols[col].mode()[0]
    var = num_cols[col].var()
    skew = num_cols[col].skew()
    kurt = num_cols[col].kurt()
    q05 = num_cols[col].quantile(0.05)
    q95 = num_cols[col].quantile(0.95)
    new_info[col] = [mode, var, skew, kurt, q05, q95]


num_info = pd.concat([num_info, new_info])

num_info
```

Добавил в основной describe дисперсию, моду, ассиметрию, эксцесс и 2 квантиля 0.05 и 0.95

```python
plt.figure(figsize=(10,8))

sns.heatmap(data=num_cols.corr(), cmap="RdBu", vmax=1, vmin=-1)

plt.figure(figsize=(10,8))

sns.barplot(data=num_cols, y = "SalePrice", x = "OverallQual", hue="SalePrice")

plt.tight_layout()
```

Решил посмотреть рапределение SalePrice, оно с явным скосом вправо, поэтому накинул log для исправления рапределения

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.histplot(df["SalePrice"], bins=30, kde=True)
plt.title("SalePrice")

plt.subplot(2, 2, 3)
sns.boxplot(x = df["SalePrice"])
plt.title("SalePrice")


plt.subplot(2, 2, 2)
sns.histplot(np.log1p(df["SalePrice"]), bins=30, kde=True)
plt.title("log(1 + SalePrice)")

plt.subplot(2, 2, 4)
sns.boxplot(x = np.log1p(df["SalePrice"]))
plt.title("log(1 + SalePrice)")


plt.tight_layout()
plt.show()
```
